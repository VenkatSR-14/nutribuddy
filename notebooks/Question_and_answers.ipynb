````python
import numpy as np

# Detecting outliers using the IQR method for numerical columns in all datasets

# Function to detect anomalies based on IQR
def detect_anomalies(df, numeric_columns):
    anomalies = {}
    for column in numeric_columns:
        Q1 = df[column].quantile(0.25)
        Q3 = df[column].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        anomalies[column] = ((df[column] < lower_bound) | (df[column] > upper_bound)).sum()
    return anomalies

# Analyzing anomalies for each dataset

# Exercise Dataset
exercise_numeric_cols = ['Calories Burn', 'Dream Weight', 'Actual Weight', 'Age', 'Duration',
                         'Heart Rate', 'BMI', 'Exercise Intensity']
exercise_anomalies = detect_anomalies(exercise_df, exercise_numeric_cols)

# User Profiles Dataset for Exercise
exercise_user_profiles_numeric_cols = ['Age', 'Preferred Intensity', 'Preferred Duration']
exercise_user_profiles_anomalies = detect_anomalies(exercise_user_profiles_df, exercise_user_profiles_numeric_cols)

# Recent Activity Dataset for Exercise
exercise_recent_activity_numeric_cols = ['Rated', 'Liked', 'Performed', 'Duration']
exercise_recent_activity_anomalies = detect_anomalies(exercise_recent_activity_df, exercise_recent_activity_numeric_cols)

# Diet Dataset
diet_numeric_cols = ['Price']
diet_anomalies = detect_anomalies(diet_df, diet_numeric_cols)

# User Profiles Dataset for Diet
diet_user_profiles_numeric_cols = []  # All columns are categorical or text-based
diet_user_profiles_anomalies = {}

# Recent Activity Dataset for Diet
diet_recent_activity_numeric_cols = ['Rated', 'Liked', 'Searched', 'Purchased']
diet_recent_activity_anomalies = detect_anomalies(recent_activity_df, diet_recent_activity_numeric_cols)

# Outputting anomalies
exercise_anomalies, exercise_user_profiles_anomalies, exercise_recent_activity_anomalies, \
diet_anomalies, diet_user_profiles_anomalies, diet_recent_activity_anomalies
````
